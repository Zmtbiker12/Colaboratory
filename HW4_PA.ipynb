```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
```

```python
g = 9.81 #m/s^2
qin0 = 0.1 # m/s
qin = (1.5)*qin0
Ac1, Ac2, Ac3 = 1, 0.5, 2 #Cross sectional area of the tanks 
Ad = np.array([0.01, 0.01, 0.01]) #Cross sectional area of the drainage hole
tend = 10000
```

```python
h0 = (qin0**2)/((Ad**2)*2*g)
```

```python
def dhdt(t, h):
  h1,h2,h2 = h
  q1,q2,q3 = Ad*(np.sqrt(2*g*h))
  return [(qin - q1)/Ac1, (qin - q2)/Ac2, (qin - q3)/Ac3]
```

```python
res = solve_ivp(dhdt, (0,tend), h0,dense_output=True)
```

```python
tplot = np.linspace(0,tend,500)
h1,h2,h3 = res.sol(tplot)
```

```python
fig = make_subplots()
fig.add_scatter(x=tplot, y=h1, name='H1', mode='lines')
fig.add_scatter(x=tplot, y=h2, name='H2', mode='lines')
fig.add_scatter(x=tplot, y=h3, name='H3', mode='lines')
```
