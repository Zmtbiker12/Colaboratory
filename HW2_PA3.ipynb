```python
import numpy as np
from plotly.subplots import make_subplots

#Importing Libraries
```

```python
R = 8.314
Tc = 647.096
Pc = 22.046e6
b = (1/8) * R * (Tc/Pc)

# R constant and T & P at the crtical point. Set B coefficient
```

```python
def vdwP(v, T):
  a = 27/64 * (R**2)*(Tc**2)/Pc
  e=0.
  s=0.
  return R*T/(v-b) - a/((v-e*b)*(v+s*b))

  # Set up a function for Van der Waals equation of state, and return the equation. 
  # a coefficient and its equation
  # b is defined outside of the fucntion becasue it is used in the volume calculation below
  # e and s are "pure numbers" and are the same for all substances. They are zero because we are using Van der Waals equation of state, and don't need them. 
```

```python
v = np.logspace(np.log10(1.01*b), -1, 1000)
# Creates an array of volumes with evenly spaced logged intervals from 1.01*b to -1 [with 1000 intervals]
```

```python
fig = make_subplots()

for T in np.arange(275, 900, 25):
  fig.add_scatter(x=v, y=vdwP(v, T), mode='lines', name=f'{T}K' )
fig.update_xaxes(type='log')
fig.update_yaxes(range = [1e-7, 1e8])
fig.update_layout(width=800, height =600)

# Make and graph the plot of P versus V with respect to the different Tempuratures above and below the critical point. 
```
